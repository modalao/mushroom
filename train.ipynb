```python
mushrooms.replace('?', pd.NA, inplace=True)
mushrooms.isnull().sum()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 创建列联表并进行卡方检验
def perform_chi2_test(mushrooms, col1, col2):
    contingency_table = pd.crosstab(mushrooms[col1], mushrooms[col2])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    return chi2, p

# 遍历每对特征变量，并判断相关性
for col1 in mushrooms.columns:
    for col2 in mushrooms.columns:
        if col1 != col2:
            chi2, p = perform_chi2_test(mushrooms, col1, col2)
            print(f"Chi-square test between {col1} and {col2}:")
            print(f"  Chi2: {chi2}")
            print(f"  P-value: {p}")
            if p < 0.05:  # 设定显著性水平为0.05
                print("  结论: 变量之间存在显著关联")
            else:
                print("  结论: 变量之间不存在显著关联")
            print()
```
